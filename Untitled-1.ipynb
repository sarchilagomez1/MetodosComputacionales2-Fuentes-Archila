```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función sin el arctan
def I_y_without_arctan(omega):
    numerador = np.sqrt(2.5e-13 + 1 / omega**2)
    denominador_parte1 = 5.625e-31
    denominador_parte2 = 2.4995e-18 / omega**2
    denominador_parte3 = 9.98e-7 / omega**4
    denominador_parte4 = 1 / omega**6
    denominador = np.sqrt(denominador_parte1 + denominador_parte2 + denominador_parte3 + denominador_parte4)
    
    return -3.75e-12 * (numerador / denominador)

# Crear un rango de valores para omega
omega_vals = np.logspace(2, 8, 500)  # omega desde 10^2 hasta 10^8
I_y_vals = I_y_without_arctan(omega_vals)

# Graficar la función
plt.figure(figsize=(8, 6))
plt.plot(omega_vals, I_y_vals, label=r'$I_y \ sin \ arctan$', color='b')
plt.xscale('log')
plt.yscale('linear')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'$I_y$')
plt.title(r'Gráfica de $I_y$ sin el término arctan')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de la fase (arctan en función de omega)
def phase_arctan(omega):
    R = 2e3  # 2 kohms
    L = 1e-3  # 1 mH
    C = 0.5e-6  # 0.5 uF
    
    # Calcular el numerador y denominador del arctan
    numerador = (R**2) / (omega**3 * L**2 * C) - 1 / (omega * C) - 4 * R**2 / (omega * L)
    denominador = 3 * R - (R**3) / (omega**2 * L**2) - 2 * R / (omega**2 * L * C)
    
    return np.arctan(numerador / denominador)

# Crear un rango de valores para omega
omega_vals = np.logspace(2, 8, 500)  # omega desde 10^2 hasta 10^8
phase_vals = phase_arctan(omega_vals)

# Graficar la función de fase (arctan)
plt.figure(figsize=(8, 6))
plt.plot(omega_vals, phase_vals, label=r'$\text{Fase} \, (arctan)$', color='r')
plt.xscale('log')
plt.yscale('linear')
plt.xlabel(r'$\omega$ (rad/s)')
plt.ylabel(r'Fase (radianes)')
plt.title(r'Gráfica de la fase (arctan) en función de $\omega$')
plt.grid(True, which="both", ls="--")
plt.legend()
plt.show()
```
