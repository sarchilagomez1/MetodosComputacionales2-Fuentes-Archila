```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def system(t,y,omega_0,drag,c):
    x,v = y 
    return np.array([v, -omega_0**2*x - drag*v - c*x**3])

sol = solve_ivp(system, y0=np.array([1.,0.]), t_span=(0,100), args=[0.5,0.05,20], max_step=0.1)
```

```python
#Plot de x vs t 
plt.plot(sol.t, sol.y[0])
```

```python
#Plot de v vs t 
plt.plot(sol.t, sol.y[1])
```

```python
#Plot de E vs t 
plt.plot(sol.t, sol.y[0]**2 *0.5**2 + sol.y[1]**2)
```
